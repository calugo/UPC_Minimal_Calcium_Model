```julia
using PyPlot
```

```julia
function Is(t,p)
    
    Iₒ,T = p
    I = 0.0
    wo = 2.0*π*10.0/T
    w = 2.0*π/T
  
    U = sin(w*t)

    (U>0.0 && (((w*t)%π)>0 &&((w*t)%π) <= wo) ) ? I=Iₒ : I=0.0 
    
    return I
end 


function Isx(t,p)
    
    Iₒ,T = p
    I = 0.0
    a = t - T*floor(t/T)
    b = 10.0
    #println(t,' ',a,' ', t - a)

    ( (a <= b) && ((t-a)>=0.0)) ? I=Iₒ : I=0.0 
    
    return I
end 


function dcai(p,u)
    grel, ka, kom, csr, co, tdiff, kim, ki, Io, To = p
    t,c,q = u
    Ic = Isx(t,[Io,To])
    dc = Ic + grel*q*(ka*c^2)*(csr - c)/(kom+ka*c^2) - (c - co)/tdiff
    dq = kim*(1-q)-ki*c*q
    
    return [dc,dq]
    
end

To = 300.0
Tr = 400.0
ki = 1.0/Tr
uₒ=[0.0,0.0]
pₒ=[0.556,12,0.12,500/1e3,0.1/1e3,2.0,ki,0.5,5.0/1000,To]
dt = 0.1

t = 0.0:dt:50*To
u = uₒ
p = pₒ

ca= zero(t)
qa = zero(t)
is =zero(t)

for (i,τ) in enumerate(t)
    x = [τ,u[1],u[2]]
  
    k1 = dcai(p,x) 
    k2 = dcai(p,[τ+dt*0.5,x[2]+0.5*dt*k1[1],x[3]+0.5*dt*k1[2]])
    k3 = dcai(p,[τ+dt*0.5,x[2]+0.5*dt*k2[1],x[3]+0.5*dt*k2[2]])
    k4 = dcai(p,[τ+dt,x[2]+dt*k3[1],x[3]+dt*k3[2]])
    u +=  (dt/6.0)*(k1+2.0*k2+2.0*k3+k4)
    ca[i] = u[1]
    qa[i] = u[2]
    is[i] = Isx(τ,[5.0/1000,To])

end
```

```julia
fig, (ax,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20,10))
ax.plot(t,qa)
ax2.plot(t,ca)
```
